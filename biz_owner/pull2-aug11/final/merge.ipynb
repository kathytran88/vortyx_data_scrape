```python
import pandas as pd
import numpy as np

df1 = pd.read_excel('NY Google API Pull 1.xlsx')
df2 = pd.read_excel('pull 2.xlsx')
```

```python
df1.shape
```

```python
for index, phone in df1['phoneNumber'].items():
    if pd.notnull(phone) and '+1' in phone:
        df1.at[index, 'phoneNumber'] = phone.replace('+1 ', '').replace('+1', '') 


for index, phone in df1['phoneNumber'].items():
    if pd.notnull(phone):
        new_phone = ''

        for character in phone:
            if character.isnumeric():
                new_phone += character 
        df1.at[index, 'phoneNumber'] = new_phone
    
df1.head()
```

```python
df1.drop_duplicates(subset='phoneNumber')
df1.shape
```

```python
for index, phone in df1['phoneNumber'].items():
    if pd.notnull(phone):
        df1.at[index, 'phoneNumber'] = '+1 ' + phone[:3] + '-' + phone[3:6] + '-' + phone[6:]

df1.head()
```

```python
del df1['Unnamed: 0']
df1.shape
```

```python
del df2['Unnamed: 0']
df2.shape
```

```python
df1.head()
```

```python
df1 = df1.rename(columns = {'placeUrl': 'placeURL'})
df1.head(2)
```
