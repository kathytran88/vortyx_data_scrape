```python
import pandas as pd
import numpy as np 
df1 = pd.read_csv('454.csv')
df2 = pd.read_excel('total2.xlsx')
```

```python
#df1 = df1.drop(columns=['status', 'subtitle'])
df1 = df1.rename(columns = {'placeUrl': 'placeURL'})
df1.head(2)
```

```python
fixphone = df1[ df1['phoneNumber'].isnull() & df1['phoneFromWebsite'].notnull()] 
fixphone
```

```python
df1.loc[ df1['phoneNumber'].isnull() & df1['phoneFromWebsite'].notnull(), 'phoneNumber' ] = df1['phoneFromWebsite']
```

```python
phonenull = df1[df1['phoneNumber'].isnull()]
phonenull
```

```python
del df1['phoneFromWebsite']
df1.shape
```

```python
del df2['Unnamed: 0']
del df2['Unnamed: 0.1']
df2.head(2)
```

```python
df2.shape
```

```python
df = pd.concat([df1, df2], ignore_index = True)
df.head()
```

```python
df.shape
```

```python
df = df.drop_duplicates(keep = 'first')
df.shape
```

```python
df.tail()
```

```python
for index, phone in df['phoneNumber'].items():
    if pd.notnull(phone) and '+1' in phone:
        # Remove the "+1" prefix and any following space
        df.at[index, 'phoneNumber'] = phone.replace('+1 ', '').replace('+1', '')

df.head()
```

```python
for index, phone in df['phoneNumber'].items():
    if pd.notnull(phone):

        new_phone = ''
        for character in phone:
            if character.isnumeric():
                new_phone += character

        df.at[index, 'phoneNumber'] = new_phone 

df.head()
```

```python
df = df.drop_duplicates(subset = 'phoneNumber')
df.shape
```

```python
for index, phone in df['phoneNumber'].items():
    if pd.notnull(phone):
        df.at[index, 'phoneNumber'] = '+1 ' + phone[:3] + '-' + phone[3:6] + '-' + phone[6:]

df.head()
```
