```python
import pandas as pd
import numpy as np 
import glob

excel_files = glob.glob('./*xlsx')
all_df = []
for file in excel_files:
    dfx = pd.read_excel(file)
    all_df.append(dfx)

df = pd.concat(all_df, ignore_index = True)
df.head()
```

```python
del df['Unnamed: 0']
df.shape
```

```python
df.to_excel('10,031 businesses.xlsx')
```
