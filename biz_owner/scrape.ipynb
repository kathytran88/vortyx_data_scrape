```python
websitedf = df2['website']
websitedf = list(websitedf)
print(websitedf)
```

```python
from selenium import webdriver 
from bs4 import BeautifulSoup 
import time

owners = []
count = 0
keywords = ['founder', 'owner', 'CEO', 'co-founder', 'president', 'title', 'name']

driver = webdriver.Chrome()

for link in websitedf:
    if pd.isnull(link):
        print(count, 'N/A')
        count += 1
        owners.append(np.nan)
    else:
        try:
            driver.get(link)
            time.sleep(10)
            page_source = driver.page_source 
            soup = BeautifulSoup(page_source, 'html.parser')

            # Find and click the "team" or "our team" link
            team_page = None
            for a_tag in soup.find_all('a', href=True):
                if 'team' in a_tag.text.lower() or 'our team' in a_tag.text.lower():
                    team_page = a_tag['href']
                    break
            
            # If "team" page not found, look for "about us" page
            if not team_page:
                for a_tag in soup.find_all('a', href=True):
                    if 'about' in a_tag.text.lower() or 'about us' in a_tag.text.lower():
                        team_page = a_tag['href']
                        break

            if team_page:
                if not team_page.startswith('http'):
                    team_page = link.rstrip('/') + '/' + team_page.lstrip('/')
                driver.get(team_page)
                time.sleep(10)
                page_source = driver.page_source
                soup = BeautifulSoup(page_source, 'html.parser')

                # Extract owner/founder names
                owner_name = None
                for tag in ['h1', 'h2', 'h3', 'h4', 'h5', 'h6', 'p', 'span']:
                    for element in soup.find_all(tag):
                        if any(keyword.lower() in element.get_text(strip=True).lower() for keyword in keywords):
                            owner_name = element.get_text(strip=True)
                            break
                    if owner_name:
                        break

                if owner_name:
                    print(count, owner_name)
                    count += 1
                    owners.append(owner_name)
                else:
                    print(count, 'N/A')
                    owners.append(np.nan)
            else:
                print(count, 'Team or about us page not found', link)
                count += 1
                owners.append(np.nan)
        except:
            print(count, 'N/A')
            count += 1
            owners.append(np.nan)

# Quit the WebDriver
driver.quit()
    
```
